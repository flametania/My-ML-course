## Построение читабельного графика на основе реальной задачи

**Задача.** Экзамен по теории вероятностей и математической статистике 251 группа механико-математического факультета сдала со следующими результатами: 2, 4, 2, 4, 3, 5, 3, 2, 4, 3, 4, 3, 2, 4, 3, 5, 4, 2, 4, 3, 4, 4, 3, 4, 5. Построить статистический ряд, полигон частот.

Создаем список оценок.

```python
lst_251 = [2, 4, 2, 4, 3, 5, 3, 2, 4, 3, 4, 3, 2, 4, 3, 5, 4, 2, 4, 3, 4, 4, 3, 4, 5]
```

Задаем координаты.

```python
# координаты по Ox - все варианты оценок
x_251 = list(set(lst_251))
# координаты по Oy пока инициализируем нулями
y_251 = [0] * (len(x_251))
```

### Статистический ряд

Вспомогательная функция для вывода статистического ряда в виде таблицы.

```python
def print_statistic_row(x, y, str_number):
  print('Статистический ряд', str_number)
  s = 23
  for i in range(s):
    print('-', end='')
  print()
  print('| z_i |', end='')
  for i in range(len(x)):
    print(f'{x[i]: >3}|', end='')
  print()
  for i in range(s):
    print('-', end='')
  print()
  print('| n_i |', end='')
  for i in range(len(y)):
    print(f'{y[i]: >3}|', end='')
  print()
  for i in range(s):
    print('-', end='')
  print()
```

Подсчет и вывод результатов.

```python
for c in lst_251:
  y_251[c - 2] += 1
print_statistic_row(x_251, y_251, '')
```

### График

На примере этого графика показано использование ряда полезных команд и параметров.

Команда `plot`:
* Заданы координаты по осям абсцисс и ординат.
* Параметр `color` регламентирует цвет графика. О том, какие существуют способы задания цветов, можно прочитать по ссылке https://matplotlib.org/stable/users/explain/colors/colors.html#colors-def.
* Параметр `marker` задает маркировку точек. Список возможных значений этого параметра представлен по ссылке https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers.
* Параметр `label` определяет метку графика.

Команда `xticks` задает значения оси абсцисс (для оси ординат используется команда `yticks`).

Команда `grid` с аргументом `True` добавляет масштабную сетку.

Команды `xlabel` и `ylabel` задают подписи к осям (абсцисс и ординат соответственно).

Команда `title` задает название графика. Параметр `loc` позволяет регулировать положение названия.

Команда `legend` показывает легенду графика.

```python
plt.plot(x_251, y_251, color='cyan', marker='o', label="251 группа")
plt.xticks(x_251)
plt.grid(True)
plt.xlabel("Оценка")
plt.ylabel("Количество студентов")
plt.title('''Полигон частот
Теория вероятностей и математическая статистика''', loc='left')
plt.legend();
```

Создадим список названий оценок.

```python
x_251_text = ["Пересдача", "Удовлетворительно", "Хорошо", "Отлично"]
```

Команде `xticks` можно подать список текстовых меток, которые нужно расположить на соответствующих местах числовых. Параметр `rotation` задает угол поворота в градусах.

```python
plt.plot(x_251, y_251, color='cyan', marker='o', label="251 группа")
plt.xticks(x_251, x_251_text, rotation=20)
plt.grid(True)
plt.xlabel("Оценка")
plt.ylabel("Количество студентов")
plt.title('''Полигон частот
Теория вероятностей и математическая статистика''', loc='left')
plt.legend();
```

## Столбчатая диаграмма.

Вынесем список цветов столбцов диаграммы.

```python
colors_of_marks = ['red', 'darkorange', 'gold', 'springgreen']
```

Команда `bar` создает столбчатую диаграмму. Используемые параметры - названия столбцов, список высот столбцов, список цветов столбцов.

```python
plt.bar(x_251, y_251, color=colors_of_marks)
plt.xticks(x_251)
plt.grid(True)
plt.xlabel("Оценка")
plt.ylabel("Количество студентов")
plt.title('''251 группа
Теория вероятностей и математическая статистика''', loc='left');
```

При использовании текстовых подписей столбцов диаграмму лучше располагать горизонтально. В библиотеке `matplotlib` для этого используется команда `barh`.

```python
plt.barh(x_251, y_251, color=colors_of_marks)
plt.yticks(x_251, x_251_text, rotation=20)
plt.grid(True)
plt.xlabel("Оценка")
plt.ylabel("Количество студентов")
plt.title('''251 группа
Теория вероятностей и математическая статистика''', loc='left');
```

## Круговая диаграмма

Команда `pie` используется для создания круговой диаграммы:
* Заданы численные значения фрагментов круговой диаграммы и подписи к ним (параметр `labels`).
* Параметр `colors` регламентирует цвета фрагментов.
* Параметр `startangle` регулирует угол поворота диаграммы (в градусах) против часовой стрелки, начиная от оси абсцисс.
* Параметр `explode` задает список смещений фрагментов (длина списка должна совпадать с количеством фрагментов).
* Параметр `autopct` отвечает за формат представления числовых значений фрагментов.

```python
plt.pie(y_251, labels=x_251_text, colors=colors_of_marks,
        startangle=90,
        explode=(0, 0, 0, 0.2),
        autopct='%1.1f%%')

plt.title('''251 группа
Теория вероятностей и математическая статистика''', loc='left');
```

## Несколько графиков на одном полотне. Сохранение графика в файл

### Два графика в одной строке

Для расположения двух или более графиков на одном полотне можно использовать команду `subplots`. Параметры - количество строк с графиками, количество столбцов и размер полотна (ширина и высота в дюймах).

Положения графиков указываются аналогично элементам массива. В случае одной строки можно указывать одну координату графика на полотне, иначе нужно две координаты. К командам, добавляющим метки и названия, спереди дописывается префикс `set_`.

Команда `suptitle` задает название всего полотна.

```python
fig, axs = plt.subplots(1, 2, figsize=(13, 6))
fig.suptitle('''251 группа
Теория вероятностей и математическая статистика''')

axs[0].grid(True)
axs[0].bar(x_251, y_251, color=colors_of_marks)
axs[0].set_xticks(x_251, x_251_text, rotation=20)
axs[0].set_xlabel("Оценка")
axs[0].set_ylabel("Количество студентов")

axs[1].pie(y_251, labels=x_251_text, colors=colors_of_marks,
        startangle=90,
        explode=(0, 0, 0, 0.2),
        autopct='%1.1f%%');
```

### Шесть графиков в квадрате $3 \times 2$

Добавим результаты экзаменов ещё двух групп.

```python
lst_241 = [2, 4, 3, 4, 3, 5, 3, 2, 4, 3, 4, 4, 4, 4, 3, 5, 5, 3, 4, 3, 4, 4, 3, 4, 5]
lst_231 = [2, 2, 3, 4, 2, 5, 3, 2, 4, 3, 5]

y_241 = [0] * len(x_251)
y_231 = [0] * len(x_251)
```

Подсчитаем количество оценок каждого типа для них.

```python
for c in lst_241:
  y_241[c - 2] += 1

for c in lst_231:
  y_231[c - 2] += 1
```

Сократим подписи к оценкам.

```python
x_251_text = ["Неуд", "Удовл", "Хор", "Отл"]
```

Отрисуем графики.

```python
fig, axs = plt.subplots(3, 2, figsize=(10, 18))
fig.suptitle('Теория вероятностей и математическая статистика')

axs[0, 0].grid(True)
axs[0, 0].bar(x_251, y_251, color=colors_of_marks)
axs[0, 0].set_xticks(x_251, x_251_text, rotation=20)
axs[0, 0].set_xlabel("Оценка")
axs[0, 0].set_ylabel("Количество студентов")
axs[0, 0].set_title("251 группа")

axs[0, 1].pie(y_251, labels=x_251_text, colors=colors_of_marks,
        startangle=90,
        explode=(0, 0, 0, 0.2),
        autopct='%1.1f%%')
axs[0, 1].set_title("251 группа");

axs[1, 0].grid(True)
axs[1, 0].bar(x_251, y_241, color=colors_of_marks)
axs[1, 0].set_xticks(x_251, x_251_text, rotation=20)
axs[1, 0].set_xlabel("Оценка")
axs[1, 0].set_ylabel("Количество студентов")
axs[1, 0].set_title("241 группа")

axs[1, 1].pie(y_241, labels=x_251_text, colors=colors_of_marks,
        startangle=90,
        explode=(0, 0, 0, 0.2),
        autopct='%1.1f%%')
axs[1, 1].set_title("241 группа");

axs[2, 0].grid(True)
axs[2, 0].bar(x_251, y_231, color=colors_of_marks)
axs[2, 0].set_xticks(x_251, x_251_text, rotation=20)
axs[2, 0].set_xlabel("Оценка")
axs[2, 0].set_ylabel("Количество студентов")
axs[2, 0].set_title("231 группа")

axs[2, 1].pie(y_231, labels=x_251_text, colors=colors_of_marks,
        startangle=90,
        explode=(0, 0, 0, 0.2),
        autopct='%1.1f%%')
axs[2, 1].set_title("231 группа");
```

**Замечание.** В группе 231 студентов вполовину меньше, чем в двух других. Однако высота столбца с четырьмя студентами почти такая же, как высота столбца десятью на предыдущем графике. Это можно отследить по шкале, но как правило, при сравнении графиков на шкалу смотрят не сразу (или не смотрят вообще). Поэтому лучше поставить на всех трех столбчатых диаграммах одинаковые границы.

Для того, чтобы задать границы числовых значений по оси ординат, можно использовать функцию `set_ylim`. Параметры - граничные значения (вещественные числа).

```python
fig, axs = plt.subplots(3, 2, figsize=(10, 18))
fig.suptitle('Теория вероятностей и математическая статистика')

axs[0, 0].grid(True)
axs[0, 0].set_ylim(0, 12)
axs[0, 0].bar(x_251, y_251, color=colors_of_marks)
axs[0, 0].set_xticks(x_251, x_251_text, rotation=20)
axs[0, 0].set_xlabel("Оценка")
axs[0, 0].set_ylabel("Количество студентов")
axs[0, 0].set_title("251 группа")

axs[0, 1].pie(y_251, labels=x_251_text, colors=colors_of_marks,
        startangle=90,
        explode=(0, 0, 0, 0.2),
        autopct='%1.1f%%')
axs[0, 1].set_title("251 группа");

axs[1, 0].grid(True)
axs[1, 0].set_ylim(0, 12)
axs[1, 0].bar(x_251, y_241, color=colors_of_marks)
axs[1, 0].set_xticks(x_251, x_251_text, rotation=20)
axs[1, 0].set_xlabel("Оценка")
axs[1, 0].set_ylabel("Количество студентов")
axs[1, 0].set_title("241 группа")

axs[1, 1].pie(y_241, labels=x_251_text, colors=colors_of_marks,
        startangle=90,
        explode=(0, 0, 0, 0.2),
        autopct='%1.1f%%')
axs[1, 1].set_title("241 группа");

axs[2, 0].grid(True)
axs[2, 0].set_ylim(0, 12)
axs[2, 0].bar(x_251, y_231, color=colors_of_marks)
axs[2, 0].set_xticks(x_251, x_251_text, rotation=20)
axs[2, 0].set_xlabel("Оценка")
axs[2, 0].set_ylabel("Количество студентов")
axs[2, 0].set_title("231 группа")

axs[2, 1].pie(y_231, labels=x_251_text, colors=colors_of_marks,
        startangle=90,
        explode=(0, 0, 0, 0.2),
        autopct='%1.1f%%')
axs[2, 1].set_title("231 группа");
```

### Использование функции `subplots_adjust`

Вспомогательная функция, по числу возвращает список и соответствующую списку подпись.

```python
def get_list_and_title(num):
  if num == 0:
    return y_251, "251 группа"
  if num == 1:
    return y_241, "241 группа"
  if num == 2:
    return y_231, "231 группа"
```

Для большого количества однотипных графиков можно использовать функцию `subplots_adjust`. Внутри цикла с помощью команды `add_subplot` задается число строк и столбцов графиков на полотне и положение текущего графика (нумерация идет от $1$ до $n$, где $n$ - число графиков на полотне).


```python
fig = plt.figure(figsize=(10, 18))
fig.subplots_adjust()

# счетчик графиков
i = 0

for i in range(6):
  ax_ = fig.add_subplot(3, 2, i + 1)
  y_group, group_title = get_list_and_title(i // 2)
  # если i четное, нужно рисовать столбчатую диаграмму
  if i % 2 == 0:
    ax_.grid(True)
    ax_.set_ylim(0, 12)
    ax_.set_xticks(x_251, x_251_text, rotation=20)
    ax_.set_xlabel("Оценка")
    ax_.set_ylabel("Количество студентов")
    ax_.bar(x_251, y_group, color=colors_of_marks)
    ax_.set_title(group_title)
  # если i нечетное, рисуем круговую диаграмму
  else:
    ax_.pie(y_group, labels=x_251_text, colors=colors_of_marks,
        startangle=90,
        explode=(0, 0, 0, 0.2),
        autopct='%1.1f%%')
    ax_.set_title(group_title);
  # каждый раз увеличиваем i на единицу
  i += 1
```
